```python
import json
a=open("record.json",'r')
txt1=a.read()
a.close()
txt=json.loads(txt1)
```

```python
print("##########INVENTORY LIST#########")
choice=1
while(choice!=5):
  print("1.Check stock\n2.Restock\n3.Remove defective from stock\n4.Add new item\n5.Exit")
  choice= int(input("Enter choice"))
  if choice==1:
    productid= str(input("Enter product id"))
    print("Current stock of "+str(txt[productid]["name"])+" is:"+str(txt[productid]["qty"]))
  elif choice==2:
    productid= str(input("Enter product id of item to be restocked"))
    restockqty= int(input("Enter number of units to be added"))
    txt[productid]['qty']+=restockqty
    print("Current stock of "+str(txt[productid]["name"])+" is:"+str(txt[productid]["qty"]))
  elif choice==3:
    productid= str(input("Enter product id of defective item"))
    defqty= int(input("Enter number of defective units "))
    txt[productid]['qty']-=defqty
    print("Current stock of "+str(txt[productid]["name"])+" is:"+str(txt[productid]["qty"]))
    if txt[productid]['qty']==0:
      print("STOCK EMPTY: RESTOCK")
    
  elif choice==4:
    productid=str(input("Enter ProductID: "))
    pname=str(input("Enter Product name: "))
    ptype=str(input("Enter Product type: "))
    price=int(input("Enter Product price/unit: "))
    quantity=int(input("Enter Product quantity in stock: "))
    add= {'name':pname,'Type':ptype,'ppu':price,'qty':quantity}
    txt[productid]=add
  elif choice==5:
    break;
    #choice2=str(input("1.Spices\n2.Condiments\n3.Grains and pulses\n4.Cakes and dairy\n5.Snacks\n6.Cleaning\n7.Soaps and Shampoos\n8.Instant noodles\n9.Beverages\n10.Chips and namkeen\nEnter category name"))
    
```

```python
print("WELCOME TO THE BAZAAR!")
choice=1
count=1
bill={}
while (choice!=2):
  choice= int(input("1.Shop for items\n2.Done shopping\nEnter choice"))
  if choice==1:
    categ=str(input("1.Spices\n2.Condiments\n3.Grains and pulses\n4.Cakes and dairy\n5.Snacks\n6.Cleaning\n7.Soaps and Shampoos\n8.Instant noodles\n9.Beverages\n10.Chips and namkeen\nEnter category name"))
    i=0
    flag=0
    for i in txt:
      if txt[str(i)]['Type']==categ:
        print("ProductID:"+i+"\tName:"+str(txt[str(i)]['name']+"\tPrice/unit:"+str(txt[str(i)]['ppu'])+"\tQty:"+str(txt[str(i)]['qty'])))
        flag=1
    if flag==0:
      print("Please enter category name accurately! It is case sensitive")
    else:
      productid=str(input("Enter product id of item you want to buy"))
      qty= int(input("Enter the number of units"))
      if qty > txt[productid]['qty']:
        print("Not enough in stock!")
      txt[productid]['qty']-=qty
      add={'bname':str(txt[productid]['name']), 'ppu':int(txt[productid]['ppu']), 'Qty':qty, 'tprice':qty*int(txt[productid]['ppu'])}
      bill[count]=add
      count+=1
  if choice==2:
    break

```
